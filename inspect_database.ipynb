# Inspection de la Base de Données plantes.db

Visualisez rapidement le contenu de votre base de données SQLite

```python
import sqlite3
import pandas as pd
from pathlib import Path

# Chemin vers la base de données
db_path = Path('assets/plantes.db')

if db_path.exists():
    print(f'✓ Base de données trouvée: {db_path}')
else:
    print(f'✗ Base de données NOT found: {db_path}')
```

```python
# Connexion à la base de données
conn = sqlite3.connect('assets/plantes.db')
cursor = conn.cursor()

# Lister toutes les tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

print('Tables dans la base de données:')
for table in tables:
    print(f'  - {table[0]}')
```

```python
# Afficher la structure et le contenu de chaque table
for table in tables:
    table_name = table[0]
    print(f'\n--- TABLE: {table_name} ---')
    
    # Lire la table avec pandas
    df = pd.read_sql_query(f'SELECT * FROM {table_name}', conn)
    
    print(f'Nombre de lignes: {len(df)}')
    print(f'Colonnes: {list(df.columns)}')
    print(f'\nAperçu des données:')
    print(df.head(10).to_string())
```

```python
# Fermer la connexion
conn.close()
print('\n✓ Connexion fermée')
```
